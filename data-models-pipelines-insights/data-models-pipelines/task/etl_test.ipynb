# Atividade Prática: Pipeline ETL de Teste
Este notebook demonstra um ciclo completo de **ETL (Extract, Transform, Load)** utilizando Python e a biblioteca Pandas. O objetivo é simular o processamento de dados de vendas de uma loja de tecnologia.

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

print("Bibliotecas importadas com sucesso!")
```

## 1. Extração (Extract)
Nesta etapa, vamos criar dados sintéticos que simulam uma extração de um banco de dados transacional ou arquivo CSV.

```python
def extract_data():
    # Simulação de dados brutos (Raw Data)
    data = {
        'id_pedido': [101, 102, 103, 104, 105, 106, 107],
        'data': ['2024-03-01', '2024-03-01', '2024-03-02', '2024-03-02', '2024-03-03', '2024-03-03', '2024-03-03'],
        'produto': ['Laptop', 'Mouse', 'Monitor', 'Teclado', 'Laptop', 'Mouse', 'Monitor'],
        'categoria': ['Hardware', 'Periférico', 'Monitor', 'Periférico', 'Hardware', 'Periférico', 'Monitor'],
        'quantidade': [1, 2, 1, 5, 2, 3, 1],
        'valor_unitario': [3500.00, 150.00, 1200.00, 200.00, 3500.00, 150.00, 1200.00],
        'cliente_id': [50, 51, 52, 53, 50, 54, 55]
    }
    df_raw = pd.DataFrame(data)
    print("Dados Extraídos:")
    display(df_raw.head())
    return df_raw

df_pedidos = extract_data()
```

## 2. Transformação (Transform)
Agora vamos limpar os dados, converter tipos e criar novas métricas (Feature Engineering).

```python
def transform_data(df):
    # 1. Converter coluna de data para o tipo datetime
    df['data'] = pd.to_datetime(df['data'])
    
    # 2. Criar coluna de Valor Total (Quantidade * Valor Unitário)
    df['valor_total'] = df['quantidade'] * df['valor_unitario']
    
    # 3. Adicionar uma coluna de Imposto Simulado (10%)
    df['imposto'] = df['valor_total'] * 0.10
    
    # 4. Adicionar coluna de Mês/Ano para análise temporal
    df['mes_referencia'] = df['data'].dt.strftime('%Y-%m')
    
    print("Dados Transformados:")
    display(df.head())
    return df

df_transformed = transform_data(df_pedidos)
```

## 3. Carregamento (Load)
Nesta fase, enviamos os dados processados para o destino final (neste caso, um resumo analítico).

```python
def load_data(df):
    # Criando um Resumo por Categoria (Visão OLAP)
    resumo_categoria = df.groupby('categoria').agg({
        'id_pedido': 'count',# Adiciona o seguinte trecho para explicar o que é id_pedido
        'valor_total': 'sum',
        'imposto': 'sum'
    }).rename(columns={'id_pedido': 'qtd_pedidos', 'valor_total': 'faturamento_total'})
    
    print("### RELATÓRIO FINAL (LOAD) ###")
    display(resumo_categoria)
    
    # Simulando salvamento em CSV
    resumo_categoria.to_csv('relatorio_final_vendas.csv')
    print("\nArquivo 'relatorio_final_vendas.csv' gerado com sucesso!")

load_data(df_transformed)
```
